```python
import pandas as pd
import os


# Load the master file created in Step 02
df = pd.read_csv('../data/processed/olist_cleaned.csv')

print("Trimming payload for Tableau performance")
# We only keep the columns required for our specific visual narrative
tableau_cols = [
    'order_id', 
    'customer_state', 
    'customer_city', 
    'delivery_delay_days', 
    'actual_delivery_days', 
    'delivery_status', 
    'review_score', 
    'total_order_value', 
    'purchase_month'
]

# Create the final lean dataframe
tableau_df = df[tableau_cols].copy()

# Drop rows that are missing review scores since that's half of our problem statement
tableau_df = tableau_df.dropna(subset=['review_score'])

# Create a final output directory
os.makedirs('../data/final', exist_ok=True)
output_path = '../data/final/tableau_ready_olist.csv'
tableau_df.to_csv(output_path, index=False)

print(f"Success! Tableau payload saved to: {output_path}")
print(f"Final Shape: {tableau_df.shape}")
```
